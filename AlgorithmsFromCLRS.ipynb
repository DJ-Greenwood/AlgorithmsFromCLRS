<h1>Insertion Sort: Overview</h1>

<b> Use Case:</b> Insertion sort is used when the number of elements is small. It can also be useful when the input list is already partially sorted, as it can adapt to the amount of sorting that is needed.

<b>Mathematical Analysis:</b>

Each iteration of insertion sort removes one element from the input data, finds the location it belongs to in the already-sorted section of the array, and inserts it there.
It repeats until no unsorted elements remain.\
<b>Time Complexity:</b>\
<b>Best Case:</b> O(n) when the array is already sorted.\
<b>Average and Worst Cases:</b> O(n^2) where n is the number of elements. This happens when the array is reverse sorted.\
<b>Space Complexity:</b> O(1), as it only requires a constant amount of additional space.

Python Implementation
Here's a straightforward Python implementation of the Insertion Sort algorithm:

~~~!
def insertion_sort(arr):
    # Traverse through 1 to len(arr)
    for i in range(1, len(arr)):
        key = arr[i]
        # Move elements of arr[0..i-1], that are greater than key,
        # to one position ahead of their current position
        j = i-1
        while j >=0 and key < arr[j] :
                arr[j + 1] = arr[j]
                j -= 1
        arr[j + 1] = key

# Example usage
arr = [12, 11, 13, 5, 6, 7]
insertion_sort(arr)
print("Sorted array is:", arr)
~~~
This function iteratively takes each element from the unsorted portion of the array and inserts it into its correct position in the sorted portion. This is done by moving the elements that are greater than the key one position ahead to make space for the key to be inserted.

```python
def insertion_sort(arr):
    # Traverse through 1 to len(arr)
    for i in range(1, len(arr)):
        key = arr[i]
        # Move elements of arr[0..i-1], that are greater than key,
        # to one position ahead of their current position
        j = i-1
        while j >=0 and key < arr[j] :
                arr[j + 1] = arr[j]
                j -= 1
        arr[j + 1] = key

# Example usage
arr = [12, 11, 13, 5, 6, 7]
print(insertion_sort(arr))
print("Sorted array is:", arr)
```

<h1>Merge Sort: Overview</h1>

<b>Use Case:</b> Merge Sort is a comparison-based sorting algorithm that divides the unsorted list into n sublists, each containing one element (a list of one element is considered sorted), then repeatedly merges sublists to produce new sorted sublists until there is only one sublist remaining. This will be the sorted list.

<b>Mathematical Analysis:</b>

<b>Divide:</b> The list is divided into two halves.\
<b>Conquer:</b> Each half is sorted recursively.\
<b>Combine:</b> The sorted halves are merged to produce a single sorted list.\
<b>Time Complexity:</b>Best, Average, and Worst Cases: O(n log n), where n is the number of elements in the list.
<b>Space Complexity:</b> O(n) due to the temporary arrays used in the merge process.

Python Implementation
Here's a Python implementation of Merge Sort:
~~~!
def merge_sort(arr):
    if len(arr) > 1:
        mid = len(arr) // 2  # Finding the mid of the array
        L = arr[:mid]  # Dividing the array elements into 2 halves
        R = arr[mid:]

        merge_sort(L)  # Sorting the first half
        merge_sort(R)  # Sorting the second half

        # Merging the sorted halves
        i = j = k = 0

        # Copy data to temp arrays L[] and R[]
        while i < len(L) and j < len(R):
            if L[i] < R[j]:
                arr[k] = L[i]
                i += 1
            else:
                arr[k] = R[j]
                j += 1
            k += 1

        # Checking if any element was left
        while i < len(L):
            arr[k] = L[i]
            i += 1
            k += 1

        while j < len(R):
            arr[k] = R[j]
            j += 1
            k += 1

# Example usage
arr = [12, 11, 13, 5, 6, 7]
merge_sort(arr)
print("Sorted array is:", arr)
~~~
This script defines a merge_sort function that recursively splits the list into halves until the sublists contain a single element, then merges those sublists to produce a sorted list. The merge process involves comparing elements of the sublists and ordering them into a new list until all elements are sorted and merged.

```python
def merge_sort(arr):
    if len(arr) > 1:
        mid = len(arr) // 2  # Finding the mid of the array
        L = arr[:mid]  # Dividing the array elements into 2 halves
        R = arr[mid:]

        merge_sort(L)  # Sorting the first half
        merge_sort(R)  # Sorting the second half

        # Merging the sorted halves
        i = j = k = 0

        # Copy data to temp arrays L[] and R[]
        while i < len(L) and j < len(R):
            if L[i] < R[j]:
                arr[k] = L[i]
                i += 1
            else:
                arr[k] = R[j]
                j += 1
            k += 1

        # Checking if any element was left
        while i < len(L):
            arr[k] = L[i]
            i += 1
            k += 1

        while j < len(R):
            arr[k] = R[j]
            j += 1
            k += 1

# Example usage
arr = [12, 11, 13, 5, 6, 7]
merge_sort(arr)
print("Sorted array is:", arr)
```

<h1>Heap Sort: Overview</h1>
<b>Use Case:</b> Heap Sort is particularly useful in scenarios where a worst-case time complexity of O(n log n) is needed regardless of the input data. Unlike Quick Sort, which can degrade to O(n^2) in its worst case, Heap Sort maintains its efficiency across all types of data distributions. It's beneficial for applications requiring guaranteed performance without external space use, as it sorts in place.

<b>Mathematical Analysis:</b>

Heap Sort works by visualizing the input list as a nearly complete binary tree. It first builds a max heap from the input data. Then, it iteratively removes the maximum element from the heap (the root of the heap), puts it into the sorted array, and re-heaps the remaining elements.
The key operations are the building of the heap and the re-heaping (or "heapify") process, which re-establishes the heap property after removal of the root.\
<b>Time Complexity:</b>

<b>All Cases (Best, Average, and Worst):</b> O(n log n), where n is the number of elements in the list. This efficiency comes from the fact that building the heap is O(n), and each of the n deletions from the heap is O(log n).\
<b>Space Complexity:</b> O(1), since the sorting is done in place.

Python Implementation
~~~!
def heapify(arr, n, i):
    largest = i  # Initialize largest as root
    l = 2 * i + 1     # left = 2*i + 1
    r = 2 * i + 2     # right = 2*i + 2

    # See if left child of root exists and is greater than root
    if l < n and arr[l] > arr[largest]:
        largest = l

    # See if right child of root exists and is greater than the largest so far
    if r < n and arr[r] > arr[largest]:
        largest = r

    # Change root, if needed
    if largest != i:
        arr[i], arr[largest] = arr[largest], arr[i]  # swap

        # Heapify the root.
        heapify(arr, n, largest)

def heapSort(arr):
    n = len(arr)

    # Build a maxheap.
    for i in range(n//2 - 1, -1, -1):
        heapify(arr, n, i)

    # One by one extract elements
    for i in range(n-1, 0, -1):
        arr[i], arr[0] = arr[0], arr[i]   # swap
        heapify(arr, i, 0)

# Example usage
arr = [12, 11, 13, 5, 6, 7]
heapSort(arr)
print("Sorted array is:", arr)
~~~
This script first builds a max heap from the unsorted input array. It then repeatedly removes the largest element from the heap (the root of the heap, which is the first element of the array) and re-heaps the remaining elements. The heapify function is key to maintaining the heap property after each removal.

```python
def heapify(arr, n, i):
    largest = i  # Initialize largest as root
    l = 2 * i + 1     # left = 2*i + 1
    r = 2 * i + 2     # right = 2*i + 2

    # See if left child of root exists and is greater than root
    if l < n and arr[l] > arr[largest]:
        largest = l

    # See if right child of root exists and is greater than the largest so far
    if r < n and arr[r] > arr[largest]:
        largest = r

    # Change root, if needed
    if largest != i:
        arr[i], arr[largest] = arr[largest], arr[i]  # swap

        # Heapify the root.
        heapify(arr, n, largest)

def heapSort(arr):
    n = len(arr)

    # Build a maxheap.
    for i in range(n//2 - 1, -1, -1):
        heapify(arr, n, i)

    # One by one extract elements
    for i in range(n-1, 0, -1):
        arr[i], arr[0] = arr[0], arr[i]   # swap
        heapify(arr, i, 0)

# Example usage
arr = [12, 11, 13, 5, 6, 7]
heapSort(arr)
print("Sorted array is:", arr)
```

<h1>Quick Sort: Overview</h1>
<b>Use Case:</b> Quick Sort is widely favored for its performance in sorting arrays and lists in-memory. It's commonly used in various software systems and libraries due to its efficient average-case performance and relatively simple implementation. The algorithm is particularly effective for large datasets, and its divide-and-conquer approach allows it to sort big arrays quickly.

<b>Mathematical Analysis:</b>

Quick Sort works by selecting a 'pivot' element from the array and partitioning the other elements into two sub-arrays, according to whether they are less than or greater than the pivot. The sub-arrays are then sorted recursively.
The efficiency of Quick Sort depends significantly on the choice of the pivot. Ideally, the pivot divides the list into two equal parts, leading to a balanced tree.\
<b>Time Complexity:</b>

<b>Best and Average Cases:</b> O(n log n), assuming the pivot is chosen well, and the array is divided into roughly equal parts.\
<b>Worst Case:</b> O(n^2), which occurs when the smallest or largest element is always chosen as the pivot, leading to unbalanced partitions. This scenario is rare, especially if the pivot selection is randomized or implemented with a strategy to ensure more balanced partitions.\
<b>Space Complexity:</b> O(log n) on average for the recursive stack space. In the worst case, it could go up to O(n), depending on the implementation.

Python Implementation
~~~!
def partition(arr, low, high):
    i = (low-1)         # index of smaller element
    pivot = arr[high]     # pivot

    for j in range(low, high):
        # If current element is smaller than or equal to pivot
        if arr[j] <= pivot:
            i = i+1
            arr[i], arr[j] = arr[j], arr[i]

    arr[i+1], arr[high] = arr[high], arr[i+1]
    return (i+1)

# The main function that implements QuickSort
def quickSort(arr, low, high):
    if len(arr) == 1:
        return arr
    if low < high:
        # pi is partitioning index, arr[p] is now at right place
        pi = partition(arr, low, high)

        # Separately sort elements before partition and after partition
        quickSort(arr, low, pi-1)
        quickSort(arr, pi+1, high)

# Example usage
arr = [10, 7, 8, 9, 1, 5]
n = len(arr)
quickSort(arr, 0, n-1)
print("Sorted array is:", arr)
~~~
This Python implementation selects the last element as the pivot for simplicity, partitions the array around that pivot, and recursively sorts the sub-arrays. This process is repeated until the entire array is sorted.

Quick Sort's efficiency and simplicity make it a popular choice for sorting, despite the potential for a worst-case scenario of O(n^2) time complexity. Advanced pivot selection methods, such as the median-of-three rule, can help mitigate this risk and ensure more consistently efficient performance.

```python
def partition(arr, low, high):
    i = (low-1)         # index of smaller element
    pivot = arr[high]     # pivot

    for j in range(low, high):
        # If current element is smaller than or equal to pivot
        if arr[j] <= pivot:
            i = i+1
            arr[i], arr[j] = arr[j], arr[i]

    arr[i+1], arr[high] = arr[high], arr[i+1]
    return (i+1)

# The main function that implements QuickSort
def quickSort(arr, low, high):
    if len(arr) == 1:
        return arr
    if low < high:
        # pi is partitioning index, arr[p] is now at right place
        pi = partition(arr, low, high)

        # Separately sort elements before partition and after partition
        quickSort(arr, low, pi-1)
        quickSort(arr, pi+1, high)

# Example usage
arr = [10, 7, 8, 9, 1, 5]
n = len(arr)
quickSort(arr, 0, n-1)
print("Sorted array is:", arr)
```

<h1>Binary Search Trees: Overview</h1>
<b>Use Case:</b> A Binary Search Tree is a fundamental data structure that supports dynamic data set operations such as search, insert, delete, min, and max in logarithmic time. BSTs are widely used in implementing databases and file systems due to their efficiency in managing sorted data.

<b>Mathematical Analysis:</b>

A BST is a binary tree where each node has a comparable key (and an associated value) and satisfies the binary search property: the key in each node is greater than all keys in the node's left subtree and less than those in its right subtree.
The efficiency of operations depends on the tree's height. In the best case, the tree is balanced, and the height is O(log n), where n is the number of nodes. In the worst case, the tree may become skewed (essentially a linked list), and operations can degrade to O(n).\
<b>Time Complexity:</b>

<b>Search, Insert, Delete:</b>
<b>Best and Average Case:</b> O(log n), assuming a balanced tree.
<b>Worst Case:</b> O(n), in a skewed tree.

Python Implementation
Here's a simple Python implementation to demonstrate how BSTs work, focusing on the insert and search functionalities.
~~~!
class Node:
    def __init__(self, key):
        self.left = None
        self.right = None
        self.val = key

# A utility function to insert a new node with the given key
def insert(root, key):
    if root is None:
        return Node(key)
    else:
        if root.val < key:
            root.right = insert(root.right, key)
        else:
            root.left = insert(root.left, key)
    return root

# A utility function to search a given key in BST
def search(root, key):
    # Base Cases: root is null or key is present at root
    if root is None or root.val == key:
        return root

    # Key is greater than root's key
    if root.val < key:
        return search(root.right, key)

    # Key is smaller than root's key
    return search(root.left, key)

# Example usage
r = Node(50)
r = insert(r, 30)
r = insert(r, 20)
r = insert(r, 40)
r = insert(r, 70)
r = insert(r, 60)
r = insert(r, 80)

# Search for a value
if search(r, 60):
    print("Found")
else:
    print("Not Found")
~~~
This code snippet demonstrates the creation of a BST, insertion of several nodes into the tree, and a search operation to find a node. The simplicity of BST operations makes it a powerful tool for data management and retrieval.

Binary Search Trees serve as the foundation for more complex structures like AVL trees, Red-Black trees, and B-trees, all of which aim to maintain tree balance and ensure operations remain efficient (O(log n)) even in the worst case.

```python
class Node:
    def __init__(self, key):
        self.left = None
        self.right = None
        self.val = key

# A utility function to insert a new node with the given key
def insert(root, key):
    if root is None:
        return Node(key)
    else:
        if root.val < key:
            root.right = insert(root.right, key)
        else:
            root.left = insert(root.left, key)
    return root

# A utility function to search a given key in BST
def search(root, key):
    # Base Cases: root is null or key is present at root
    if root is None or root.val == key:
        return root

    # Key is greater than root's key
    if root.val < key:
        return search(root.right, key)

    # Key is smaller than root's key
    return search(root.left, key)

# Example usage
r = Node(50)
r = insert(r, 30)
r = insert(r, 20)
r = insert(r, 40)
r = insert(r, 70)
r = insert(r, 60)
r = insert(r, 80)

# Search for a value
if search(r, 60):
    print("Found")
else:
    print("Not Found")
```

<h1>AVL Trees: Overview</h1>
<b>Use Case:</b> AVL trees are used in situations where frequent insertions and deletions are expected and where balanced search time is critical. Due to their self-balancing nature, AVL trees ensure that the depth of the tree remains O(log n), making them highly efficient for dictionary and priority queue implementations, among other applications.

<b>Mathematical Analysis:</b>

Each node in an AVL tree stores a balance factor, which is the height difference between its left and right subtrees. An AVL tree requires the balance factor to be -1, 0, or +1 for every node.
Balancing is achieved through rotations during insertions and deletions when the balance factor of any node becomes less than -1 or greater than +1.
Time Complexity:

Search: O(log n) since AVL trees are always balanced.
Insertion and Deletion: O(log n), as the tree remains balanced, but these operations may require rebalancing through rotations.
Python Implementation
Here's a simplified Python implementation of an AVL Tree focusing on the insertion operation and the balance maintenance through rotations.
~~~!
class TreeNode(object):
    def __init__(self, key):
        self.key = key
        self.left = None
        self.right = None
        self.height = 1

class AVLTree(object):
    def insert(self, root, key):
        # Step 1 - Perform normal BST
        if not root:
            return TreeNode(key)
        elif key < root.key:
            root.left = self.insert(root.left, key)
        else:
            root.right = self.insert(root.right, key)

        # Step 2 - Update the height of the ancestor node
        root.height = 1 + max(self.getHeight(root.left),
                              self.getHeight(root.right))

        # Step 3 - Get the balance factor
        balance = self.getBalance(root)

        # Step 4 - If the node is unbalanced, then try out the 4 cases
        # Case 1 - Left Left
        if balance > 1 and key < root.left.key:
            return self.rightRotate(root)

        # Case 2 - Right Right
        if balance < -1 and key > root.right.key:
            return self.leftRotate(root)

        # Case 3 - Left Right
        if balance > 1 and key > root.left.key:
            root.left = self.leftRotate(root.left)
            return self.rightRotate(root)

        # Case 4 - Right Left
        if balance < -1 and key < root.right.key:
            root.right = self.rightRotate(root.right)
            return self.leftRotate(root)

        return root

    def getHeight(self, root):
        if not root:
            return 0
        return root.height

    def getBalance(self, root):
        if not root:
            return 0
        return self.getHeight(root.left) - self.getHeight(root.right)

    def leftRotate(self, z):
        y = z.right
        T2 = y.left
        y.left = z
        z.right = T2
        z.height = 1 + max(self.getHeight(z.left),
                           self.getHeight(z.right))
        y.height = 1 + max(self.getHeight(y.left),
                           self.getHeight(y.right))
        return y

    def rightRotate(self, y):
        x = y.left
        T2 = x.right
        x.right = y
        y.left = T2
        y.height = 1 + max(self.getHeight(y.left),
                           self.getHeight(y.right))
        x.height = 1 + max(self.getHeight(x.left),
                           self.getHeight(x.right))
        return x

# Example usage

root = None
nums = [10, 20, 30, 40, 50, 25]
for num in nums:
    root = avl_tree.insert(root, num)
    print("Root:", root.key)
~~~

# The AVL tree will be balanced automatically.
This example demonstrates inserting nodes into an AVL tree while maintaining its balance through rotations. There are four types of rotations to consider: left-left, right-right, left-right, and right-left, depending on the balance factor and the side of the insertion.

AVL Trees ensure that the tree remains balanced with each insertion or deletion, which maintains the efficiency of search, insert, and delete operations at O(log n).

```python
class TreeNode(object):
    def __init__(self, key):
        self.key = key
        self.left = None
        self.right = None
        self.height = 1

class AVLTree(object):
    def insert(self, root, key):
        # Step 1 - Perform normal BST
        if not root:
            return TreeNode(key)
        elif key < root.key:
            root.left = self.insert(root.left, key)
        else:
            root.right = self.insert(root.right, key)

        # Step 2 - Update the height of the ancestor node
        root.height = 1 + max(self.getHeight(root.left),
                              self.getHeight(root.right))

        # Step 3 - Get the balance factor
        balance = self.getBalance(root)

        # Step 4 - If the node is unbalanced, then try out the 4 cases
        # Case 1 - Left Left
        if balance > 1 and key < root.left.key:
            return self.rightRotate(root)

        # Case 2 - Right Right
        if balance < -1 and key > root.right.key:
            return self.leftRotate(root)

        # Case 3 - Left Right
        if balance > 1 and key > root.left.key:
            root.left = self.leftRotate(root.left)
            return self.rightRotate(root)

        # Case 4 - Right Left
        if balance < -1 and key < root.right.key:
            root.right = self.rightRotate(root.right)
            return self.leftRotate(root)

        return root

    def getHeight(self, root):
        if not root:
            return 0
        return root.height

    def getBalance(self, root):
        if not root:
            return 0
        return self.getHeight(root.left) - self.getHeight(root.right)

    def leftRotate(self, z):
        y = z.right
        T2 = y.left
        y.left = z
        z.right = T2
        z.height = 1 + max(self.getHeight(z.left),
                           self.getHeight(z.right))
        y.height = 1 + max(self.getHeight(y.left),
                           self.getHeight(y.right))
        return y

    def rightRotate(self, y):
        x = y.left
        T2 = x.right
        x.right = y
        y.left = T2
        y.height = 1 + max(self.getHeight(y.left),
                           self.getHeight(y.right))
        x.height = 1 + max(self.getHeight(x.left),
                           self.getHeight(x.right))
        return x

# Example usage
avl_tree = AVLTree()
root = None
nums = [10, 20, 30, 40, 50, 25]
for num in nums:
    root = avl_tree.insert(root, num)
    print("Root:", root.key)
```

<h1>Breadth-First Search (BFS): Overview</h1>
<b>Use Case:</b> BFS is widely used for finding the shortest path on unweighted graphs, whether in networking, finding connected components in undirected graphs, or for algorithmic puzzles and games that involve exploring all possible states. Its ability to traverse level by level makes it perfectly suited for algorithms that need to process nodes in layers or determine the minimum number of edges to reach a node from a given source.

<b>Mathematical Analysis:</b>

BFS explores all neighbors of a node before moving on to their neighbors, ensuring that all nodes at a distance 
<i>d</i> from the starting point are explored before those at a distance 
<i>d</i>+1.

It uses a queue to keep track of the nodes that need to be explored, ensuring that exploration proceeds in a first-in-first-out manner.\
<b>Time Complexity:</b>

<b>All Cases:</b> O(V + E), where 

V is the number of vertices and 

E is the number of edges in the graph. This accounts for every vertex being inserted into the queue once and every edge being explored in the case of an adjacency list representation.
Python Implementation
Let's implement BFS in Python, using an adjacency list to represent the graph:
~~~!
from collections import deque

# This function implements the BFS algorithm
def bfs(graph, start):
    visited = set() # Set to keep track of visited nodes.
    queue = deque([start]) # Create a queue for BFS

    visited.add(start)
    while queue:
        vertex = queue.popleft() # Pop a vertex from queue
        print(str(vertex) + " ", end="")

        # Get all adjacent vertices of the dequeued vertex
        # If an adjacent has not been visited, mark it visited and enqueue it
        for neighbour in graph[vertex]:
            if neighbour not in visited:
                visited.add(neighbour)
                queue.append(neighbour)

# Example usage
# Define a graph as an adjacency list
graph = {0: [1, 2], 1: [2], 2: [3], 3: [1, 2]}
print("BFS starting from vertex 2:")
bfs(graph, 2)
print(bfs(graph,2))
~~~
This code demonstrates how BFS works by traversing the graph from a specified start node and visiting all nodes in a breadth-first manner.

<h1>Depth-First Search (DFS): Overview</h1>
<b>Use Case:</b> DFS is utilized in various applications, including finding connected components in a graph, topological sorting in directed graphs, solving puzzles and mazes, and in algorithms for searching for solutions in a game or puzzle space. It's also foundational for algorithms that require visiting every vertex and edge of a graph in a methodical way.

<b>Mathematical Analysis:</b>

DFS starts at a selected node (root) and explores as far as possible along each branch before backtracking.
It uses a recursive approach or an explicit stack data structure to keep track of the path it's exploring.
Time Complexity:

All Cases: O(V + E) for an adjacency list representation of the graph, where V is the number of vertices and E is the number of edges. This is because each vertex is explored once, and each edge is considered twice (once for each direction).
Python Implementation
Here's a Python implementation of DFS using recursion:
~~~!
# This class represents a directed graph using adjacency list representation

<h1>Dijkstra's Algorithm: Overview</h1>
<b>Use Case:</b> Dijkstra's Algorithm is crucial for scenarios that require efficient route planning and network flow optimization. It can determine the shortest path from a point A to point B on maps, optimize routing decisions in telecommunications networks, and solve various optimization problems in logistics and transportation.

<b>Mathematical Analysis:</b>

The algorithm maintains a set of nodes whose shortest distance from the source is already known and a set of nodes whose shortest distance is not yet determined. Initially, the distance to the source is set to zero and to infinity for all other nodes.
At every step, it selects the node with the minimum distance from the source from the set of unsettled nodes, updates the distance for its neighbors, and moves it to the set of settled nodes. This process repeats until all nodes are settled.
Time Complexity:

All Cases: Without a priority queue, O(V^2), where V is the number of vertices. With a binary heap (min-priority queue), the complexity improves to O((V+E) log V), and with a Fibonacci heap, it can achieve O(E + V log V).
Python Implementation
Here's a simplified Python implementation of Dijkstra's Algorithm using a priority queue to manage the vertices:
~~~!
import heapq  # heapq is a binary heap implementation usually used as a priority queue

class Graph:
    def __init__(self, vertices):
        self.V = vertices
        self.graph = [[] for i in range(vertices)]

    def addEdge(self, u, v, w):
        self.graph[u].append((v, w))

    # Utility function to find the vertex with
    # minimum distance value, from the set of vertices
    # not yet included in shortest path tree
    def dijkstra(self, src):
        dist = [float("inf")] * self.V
        dist[src] = 0
        pq = [(0, src)]  # Priority queue to store (distance, vertex)

        while pq:
            (d, u) = heapq.heappop(pq)

            # 'd' is ignored here because it's already in 'dist'
            for neighbor in self.graph[u]:
                v, weight = neighbor
                if dist[v] > dist[u] + weight:
                    dist[v] = dist[u] + weight
                    heapq.heappush(pq, (dist[v], v))

        print("Vertex Distance from Source")
        for i in range(self.V):
            print(f"Vertex {i}\tDistance {dist[i]}")

<h1>Bellman-Ford Algorithm: Overview</h1>
<b>Use Case:</b> The Bellman-Ford algorithm is particularly useful in networks where weights may change dynamically or could initially be negative, requiring the algorithm to detect and report negative weight cycles. It's used in routing protocols like RIP (Routing Information Protocol) and in systems that need to adjust to changing cost metrics robustly.

<b>Mathematical Analysis:</b>

* The algorithm relaxes all the edges Vâˆ’1 times (where V is the number of vertices). Relaxation is the process of updating the distance to a vertex if a shorter path is found.
* After Vâˆ’1 iterations, if any distance can be minimized further, a negative weight cycle exists in the graph.
<b>Time Complexity:</b>

* <b>All Cases:</b> O(V*E), where V is the number of vertices and E is the number of edges. This reflects the algorithm's iterative nature, checking all edges in each iteration.
Python Implementation
Here's a Python implementation of the Bellman-Ford algorithm:
~~~!
class Graph:
    def __init__(self, vertices):
        self.V = vertices   # No. of vertices
        self.graph = []     # Default dictionary to store the graph

    def addEdge(self, u, v, w):
        self.graph.append([u, v, w])

    # Prints the distance from source to all other vertices
    def bellmanFord(self, src):
        dist = [float("Inf")] * self.V
        dist[src] = 0

        # Relax all edges |V| - 1 times
        for _ in range(self.V - 1):
            for u, v, w in self.graph:
                if dist[u] != float("Inf") and dist[u] + w < dist[v]:
                    dist[v] = dist[u] + w

        # Check for negative-weight cycles
        for u, v, w in self.graph:
            if dist[u] != float("Inf") and dist[u] + w < dist[v]:
                print("Graph contains negative weight cycle")
                return

        # Print all distance
        print("Vertex Distance from Source")
        for i in range(self.V):
            print(f"Vertex {i}\tDistance {dist[i]}")

# Example usage
g = Graph(5)

<h1>Huffman Coding: Overview</h1>
<b>Use Case:</b> Huffman Coding is used in compression technologies, including the JPEG image format and the MP3 audio format. It's particularly effective for compressing data where certain characters or symbols occur more frequently than others, as it assigns shorter codes to more frequent characters and longer codes to less frequent ones.

<b>Mathematical Analysis:</b>

* Huffman Coding builds a binary tree where each leaf node represents a character and its frequency in the dataset. The tree is constructed in a bottom-up approach, starting with the least frequent characters.
* The path from the root to a leaf represents the binary code for the corresponding character, with left edges representing a bit of 0 and right edges a bit of 1. The goal is to minimize the overall length of the encoded data by reducing the length of codes for more common characters.
<b>Time Complexity:</b>

* The construction of the Huffman tree is O(n log n), where n is the number of characters or symbols. This is because the algorithm typically uses a priority queue to efficiently select the two nodes with the lowest frequency during each step of the tree construction.
Python Implementation
Here's an implementation of Huffman Coding in Python:
~~~!
import heapq
import collections

class Node:
    def __init__(self, char, freq):
        self.char = char
        self.freq = freq
        self.left = None
        self.right = None

    # Defining comparators less_than and equals
    def __lt__(self, other):
        return self.freq < other.freq

# Function to encode the characters
def encode(root, string, huffman_code):
    if not root:
        return

    # Found a leaf node
    if not root.left and not root.right:
        huffman_code[root.char] = string if string != '' else '1'

    encode(root.left, string + '0', huffman_code)
    encode(root.right, string + '1', huffman_code)

# Function to build the Huffman tree
def build_huffman_tree(text):
    if not text:
        return

    # Count frequency of appearance of each character
    # and store it in a dictionary
    frequency = collections.Counter(text)
    priority_queue = [Node(char, freq) for char, freq in frequency.items()]
    heapq.heapify(priority_queue)

    while len(priority_queue) > 1:
        # Remove the two nodes of highest priority
        # (lowest frequency) from the queue
        left = heapq.heappop(priority_queue)
        right = heapq.heappop(priority_queue)

        # Create a new internal node with these two nodes
        # as children and with frequency equal to the sum
        # of the two nodes' frequencies. Add the new node
        # to the priority queue.
        merged = Node(None, left.freq + right.freq)
        merged.left = left
        merged.right = right

        heapq.heappush(priority_queue, merged)

    # The remaining node is the root node and the tree is complete.
    root = priority_queue[0]
    huffman_code = {}
    encode(root, '', huffman_code)
    return huffman_code

# Example usage
text = "this is an example for huffman encoding"
huffman_code = build_huffman_tree(text)
print("Huffman Codes are :", huffman_code)
print("Original string was :", text)
~~~
This script builds a Huffman tree for a given text and then encodes the text using the generated Huffman codes. Each character in the text is assigned a unique binary code, with more frequent characters having shorter codes.

Huffman Coding is a classic example of a greedy algorithm that provides an optimal solution for a specific problem, in this case, data compression.
